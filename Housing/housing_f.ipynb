# Treating Null Values

First we split the columns on the basis of the data type to either numerical and object types.

```python
#Empty lists to store the various differentiated column names
o=[]
n=[]
```

```python
#Looping to store the columns names into the list
for i in df.columns:
    if((df[i].dtype)=='object'):
        o.append(i)
    else:
        n.append(i)
```

We treat the null values present in the object type variables. We know that the null values are present in the dataset where the following requirement is not present. Since it is classification type we can replace the null values with the appropriate option.

```python
df['Alley'].fillna('No alley',inplace=True)
df['MasVnrType'].fillna('None',inplace=True)
df['BsmtQual'].fillna('No Basement',inplace=True)
df['BsmtCond'].fillna('No Basement',inplace=True)
df['BsmtExposure'].fillna('No Basement',inplace=True)
df['BsmtFinType1'].fillna('No Basement',inplace=True)
df['BsmtFinType2'].fillna('No Basement',inplace=True)
df['FireplaceQu'].fillna('No Fireplace',inplace=True)
df['GarageType'].fillna('No Garage',inplace=True)
df['GarageFinish'].fillna('No Garage',inplace=True)
df['GarageQual'].fillna('No Garage',inplace=True)
df['GarageCond'].fillna('No Garage',inplace=True)
df['PoolQC'].fillna('No Pool',inplace=True)
df['Fence'].fillna('No Fence',inplace=True)
df['MiscFeature'].fillna('None',inplace=True)
```

Similarly we do it for the test data

```python
test['Alley'].fillna('No alley',inplace=True)
test['MasVnrType'].fillna('None',inplace=True)
test['BsmtQual'].fillna('No Basement',inplace=True)
test['BsmtCond'].fillna('No Basement',inplace=True)
test['BsmtExposure'].fillna('No Basement',inplace=True)
test['BsmtFinType1'].fillna('No Basement',inplace=True)
test['BsmtFinType2'].fillna('No Basement',inplace=True)
test['FireplaceQu'].fillna('No Fireplace',inplace=True)
test['GarageType'].fillna('No Garage',inplace=True)
test['GarageFinish'].fillna('No Garage',inplace=True)
test['GarageQual'].fillna('No Garage',inplace=True)
test['GarageCond'].fillna('No Garage',inplace=True)
test['PoolQC'].fillna('No Pool',inplace=True)
test['Fence'].fillna('No Fence',inplace=True)
test['MiscFeature'].fillna('None',inplace=True)
```

Now, treating the null values in the numerical type columns. We do it using looping method.

```python
for i in n:
    if(i=='LotFrontage'):
        df[i].fillna(df[i].mean(),inplace=True)
        test[i].fillna(test[i].mean(),inplace=True)
    elif(i=='GarageYrBlt'):
        df[i].fillna(df['YearBuilt'][df['GarageYrBlt'].isnull()],inplace=True)
        test[i].fillna(test['YearBuilt'][test['GarageYrBlt'].isnull()],inplace=True)
    elif(i=='SalePrice'):
        continue
    else:
        df[i].fillna(0,inplace=True)
        test[i].fillna(0,inplace=True)
```

# Data Preparation

First we will concat the test and train data

```python
work=pd.concat((df, test),axis=0)
```

```python
work.isnull().sum().sum() #Checking for any null values
```

Encoding the data. We use label encoder for the task

```python
le = LabelEncoder()
for i in o:
    work[i]=le.fit_transform(work[i])
```

Now we split the data back to the test and train data set

```python
# The new train data
ndf = work.iloc[:1168,:]
```

```python
# The new test data
ntest = work.iloc[1168:,:]
```

We drop the target column from the test data that formed in the concat process

```python
ntest = ntest.drop(columns='SalePrice',axis=1)
```

# Feature Selection

```python
pd.set_option('display.max_rows',None)
ndf.corr()['SalePrice'].sort_values()
```

We see that there are a lot of columns that does not produce a required correlation to the target variable or adds very less value. Hence we remove these features for the model prediction

```python
#Creating matrix for least correlation variables
rm = (ndf.corr()['SalePrice'][(ndf.corr()['SalePrice']).abs()<0.1].index.to_list())
```

We drop the least correlation columns from both test and train data

```python
ndf2 = ndf.drop(columns=rm)
ntest2 = ntest.drop(columns=rm)
```

```python
ndf2.shape
```

```python
ntest2.shape
```

Also we remove the column 'Utilities' as in the training data set we onlyt have a single unique value which therefore does not weigh in the model prediction

```python
ndf2 = ndf2.drop(columns='Utilities')
ntest2 = ntest2.drop(columns='Utilities')
```

# Skewness

Now we treat for skewness in the data using the threshold of 4

```python
skewness_cols = ndf2.skew().abs().sort_values(ascending=False)[ndf2.skew().abs().sort_values(ascending=False)>4].index.tolist()
```

```python
skewness_cols
```

```python
ndf2.skew()
```

Treating for skewness using the log1p method

```python
for i in skewness_cols:
    ndf2[i] = np.log1p(ndf2[i])
```

```python
ndf2.skew()
```

# Scaling

We split the train data into target and feature variables before scaling is done

```python
x = ndf2.drop(columns='SalePrice') #Feature Variables
y = ndf2['SalePrice']              #Target Variables
```

Scaling data using Standard Scaler method. We use the train data to train the scaler

```python
sc= StandardScaler()
sc.fit(x)
x = sc.transform(x)
ntest2 = sc.transform(ntest2)
```

# Model Training

```python
model = XGBRegressor(n_estimators=1000, max_depth=7, eta=0.1, subsample=0.7, colsample_bytree=0.8)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.22, random_state=100)
```

```python
model.fit(x_train,y_train, eval_set=[(x_test,y_test)], eval_metric='rmse', verbose=200, early_stopping_rounds=1000)
```

```python
y_pred = model.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

# Predicting

```python
final_predictions = model.predict(ntest2)
```

```python
pd.Series(final_predictions).to_csv('result.csv')
```
