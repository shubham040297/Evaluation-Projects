```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

# Scraping

```python
# Calling the webdriver exe file
driver = webdriver.Chrome(r"C:\Users\Shubham\Desktop\Projects\FlipRobo\chromedriver.exe")
```

```python
# Defining empty arrays for the data to be scraped
airline = []
date = []
source = []
dest = []
dep_time = []
arr_time = []
duration = []
price = []
stops = []
```

```python
# Calling the url in the webdriver
driver.get('https://flight.yatra.com/air-search-ui/dom2/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=BOM&destinationCountry=IN&flight_depart_date=04%2F05%2F2022&ADT=1&CHD=0&INF=0&class=Economy&source=fresco-home&unqvaldesktop=919161197782')
element = WebDriverWait(driver, 60).until(EC.presence_of_element_located((By.XPATH , '//span[@class="i-b text ellipsis"]')))

#Loading Screen to show all flights for the day
while True:
    load = driver.find_element_by_xpath('//div[@class="yatraSecure"]')
    p = load.location_once_scrolled_into_view
    tag = driver.find_elements_by_xpath('//span[@class="i-b text ellipsis"]')
    j = len(tag)
    if(j>150):
        break

#Scraping airline
tag1 = driver.find_elements_by_xpath('//span[@class="i-b text ellipsis"]')
for i in tag1:
    airline.append(i.text)
    date.append('4 May')

#Scraping source
tag3 = driver.find_elements_by_xpath('//div[@class="i-b col-4 no-wrap text-right dtime col-3"]/p[1]')
for i in tag3:
    source.append(i.text)

#Scraping destination
tag4 = driver.find_elements_by_xpath('//div[@class="i-b pdd-0 text-left atime col-5"]/p[2]')
for i in tag4:
    dest.append(i.text)

#Scraping departure time
tag5 = driver.find_elements_by_xpath('//div[@class="i-b col-4 no-wrap text-right dtime col-3"]/div[1]/div[1]')
for i in tag5:
    dep_time.append(i.text)

#Scraping arrival time
tag6 = driver.find_elements_by_xpath('//div[@class="i-b pdd-0 text-left atime col-5"]/p[1]')
for i in tag6:
    arr_time.append(i.text)

#Scraping duration
tag7 = driver.find_elements_by_xpath('//p[@class="fs-12 bold du mb-2"]')
for i in tag7:
    duration.append(i.text)

#Scraping stops
tag8 = driver.find_elements_by_xpath('//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]/span[1]')
for i in tag8:
    stops.append(i.text)

#Scraping price
tag9 = driver.find_elements_by_xpath('//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')
for i in tag9:
    price.append(i.text)
```
